```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("/content/Iris.csv")
df
```

```python
features=["SepalLengthCm",'SepalWidthCm','PetalLengthCm','PetalWidthCm']
X=df[features]
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(X)
clusters
```

```python
centroids = pd.DataFrame(X).groupby(clusters).mean().values
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X['SepalLengthCm'], X['SepalWidthCm'],c=clusters, cmap='viridis', edgecolors='k', s=50)
plt.title('DBSCAN Clustering of iris')
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X['PetalLengthCm'],X['PetalWidthCm'],c=clusters, cmap='viridis', edgecolors='k', s=50)
plt.title('DBSCAN Clustering of IRIS')
plt.xlabel('petalLengthCm')
plt.ylabel('petalWidthCm')
plt.show()
```
