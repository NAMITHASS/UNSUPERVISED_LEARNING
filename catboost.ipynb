```python
x=df_train.drop("ACTION",axis=1)
y=df_train["ACTION"]
```

```python
from sklearn.model_selection import train_test_split
X_train_sub, X_validation_sub, y_train_sub, y_validation_sub=train_test_split(x,y,random_state=0)
```

```python
from catboost import CatBoostClassifier
from sklearn.metrics import classification_report
```

```python
model=CatBoostClassifier()
```

```python
learning_rates = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
for learning_rate in learning_rates:
   cat = CatBoostClassifier(n_estimators=20, learning_rate =learning_rate, max_depth = 2, random_state = 0)
   cat.fit(X_train_sub, y_train_sub)
   #print("Learning rate: ", learning_rate)
   print("Accuracy score (training): {0:.3f}".format(cat.score(X_train_sub, y_train_sub)))
   print("Accuracy score (validation): {0:.3f}".format(cat.score(X_validation_sub, y_validation_sub)))
   print()
```
